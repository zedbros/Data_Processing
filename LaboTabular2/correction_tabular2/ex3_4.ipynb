```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

sns.set(color_codes=True)
```

```python
df = pd.read_csv('../data/iris.csv')
df.head(5)
```

```python
# a) Make a scatterplot matrix of the entire dataset with scatter-plots (pair-wise) and density estimations (univariate)
df_renamed = df.rename(columns={'sepal.length': 'Sepal Length (cm)', 'sepal.width': 'Sepal Width (cm)', 'petal.length': 'Petal Length (cm)', 'petal.width': 'Petal Width (cm)'})
g = sns.pairplot(df_renamed, hue='variety')
g.fig.suptitle("Pair-wise Relationships and Univariate Distributions in Iris Dataset", y=1.02)

```

```python
# b) Make a parallel coordinates plot 
df_renamed = df.rename(columns={'sepal.length': 'Sepal Length', 'sepal.width': 'Sepal Width', 'petal.length': 'Petal Length', 'petal.width': 'Petal Width'})
pd.plotting.parallel_coordinates(df_renamed,'variety',color=('red', 'green', 'blue'), alpha=0.2)
plt.title('Parallel Coordinates Plots of Iris Dataset')
plt.ylabel('Size (cm)')
```

```python
# c) Plot a heatmap of correlation of all features
features_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]
# sns.heatmap(df_renamed[features_names].corr())
sns.heatmap(df_renamed[features_names].corr(),annot=True,cmap="crest",vmin=-1,vmax=1)
plt.title('Correlation Heatmap of Iris Features')
plt.xticks(rotation=20) 
```
