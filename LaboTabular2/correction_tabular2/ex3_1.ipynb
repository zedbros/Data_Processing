```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

# Activating seaborn’s “default” theme
sns.set()
```

```python
df = pd.read_csv('../data/iris.csv')
df.describe()
```

```python
# a) Make a barplot representing, for each variety, the counts of sepal lengths greater than the overall mean sepal length
mean = df['sepal.length'].mean()
# Filter DataFrame and group by variety
filtered_df = df.loc[df['sepal.length'] > mean]

# Count within groups
grouped_df = filtered_df.groupby('variety')['sepal.length'].count()
# or
# grouped_df =  filtered_df['variety'].value_counts()

# Comment: Setosa has all sepal lengths < mean so not represented on the plot. We can add it to the graph with this:
# grouped_df =  filtered_df['variety'].value_counts().reindex(df['variety'].unique(), fill_value=0)

# Create bar plot
plt.figure(figsize=(4, 3))
sns.barplot(data=grouped_df)
# Or similar with df.plot:
# grouped_df.plot.bar(rot=0)

# Adding title and labels
plt.title(f"Count of Sepal Lengths > {mean:.2f} cm by Variety")
plt.xlabel('Variety')
plt.ylabel('Count')


```

```python
# b) Make a bar plot of the mean sepal length by variety, with error bars (standard deviation)
means = df.groupby('variety')['sepal.length'].mean()
stds = df.groupby('variety')['sepal.length'].std()

# Plotting the bar chart with error bars
means.plot.bar(yerr=stds, capsize=5, rot=0)

# # Adding labels and title
plt.xlabel('Variety')
plt.ylabel('Mean Sepal Length (cm)')
plt.title('Mean Sepal Length with Error Bars (std) by Variety')
```
