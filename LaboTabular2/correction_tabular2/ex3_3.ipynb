```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

sns.set()
```

```python
df = pd.read_csv('../data/iris.csv')
df.head()
```

```python
# a) Plot a single histogram of sepal lengths (all varieties together) 
df['petal.length'].plot.hist(title='Histogram of Sepal Lengths across all Varieties')

# # Adding label
plt.xlabel('Petal length (cm)')
```

```python
# b) Plot 4 histograms (1 per feature), all varieties together. Once with 4 subplots, once overlaid on the same plot.
features = ['sepal.length','sepal.width', 'petal.length', 'petal.width']
df[features].plot.hist(subplots=True,sharey=True,figsize=(6,8),title='Histograms of Sizes')
# Adding label
plt.xlabel('Size (cm)')
```

```python
# b) All 4 histograms of features, all varieties, overlaid on a single plot OK
features = ['sepal.length','sepal.width', 'petal.length', 'petal.width']
df[features].plot.hist(subplots=False, alpha=0.6, title='Overlaid Histograms of Sizes') # alpha makes them semi-transparent
# # Adding label
plt.xlabel('Size (cm)')
```

```python
# b) Same with seaborn
features = ['sepal.length','sepal.width', 'petal.length', 'petal.width']
sns.histplot(df[features], alpha=0.6)
# Adding label
plt.title('Overlaid Histograms of Sizes')
plt.xlabel('Size (cm)')
```

```python
# b) Other method for overlaying histograms
for f in features:
    plt.hist(df[f], range=(0,8), bins=15, alpha=0.8, label=f)

plt.xlabel('Size (cm)')
plt.ylabel('Frequency')
plt.title('Overlaid Histograms of Sizes')
plt.legend()
```

```python
# c) Plot the histograms of petal lengths grouped by variety (3 histograms). Once with 3 subplots, once overlaid on the same plot. 
df.plot.hist(column='petal.length',by='variety',title="Histograms of Petal Lengths by Variety", figsize=(6,7), sharex=True,sharey=True,legend=False)
plt.xlabel('Size (cm)')
plt.plot()
```

```python
# c) overlaid histograms of petal lengths grouped by variety
varieties = df['variety'].unique()
for v in varieties:
    df[df['variety'] == v]['petal.length'].plot.hist(alpha=0.8, label=f'Variety {v}')

# Add labels and legend
plt.title('Overlaid Histograms of Petal Lengths by Variety')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.plot()
```

```python
# c) Simpler with seaborn
sns.histplot(df,x='petal.length', binwidth=0.2, hue='variety',stat='frequency')
plt.title('Overlaid Histograms of Petal Lengths by Variety')
plt.xlabel('Petal Length (cm)')
```

```python
# d) Plot the KDE of sepal lengths distributions for each variety (overlaid). Compare with the histograms in c)
sns.kdeplot(df,x='petal.length', fill=True, hue='variety')
plt.title('Overlaid KDEs of Petal Lengths by Variety')
plt.xlabel('Petal Length (cm)')
```

```python
sns.kdeplot(df,x='petal.length', hue='variety')
```

```python
# e) Plot the boxplots, for the 3 varieties, of sepal lengths and petal lengths (i.e. 6 boxplots)
fig, ax = plt.subplots(1,2)

# Boxplot for 'sepal.length'
df.boxplot(column='sepal.length', by='variety', ax=ax[0])
ax[0].set_ylabel('Length (cm)')
ax[0].set_title('Sepal Length')


# Boxplot for 'petal.length'
df.boxplot(column='petal.length', by='variety', ax=ax[1])
ax[1].set_ylabel('Length (cm)')
ax[1].set_title('Petal Length')

# Adjust layout for better spacing
plt.tight_layout()
plt.suptitle('Boxplots of Sepal and Petal Lengths by Variety')
```

```python
# e) Or with seaborn
fig, ax = plt.subplots(1,2)
# Boxplot for 'sepal.length'
sns.boxplot(x='variety', y='sepal.length', data=df, width=0.5, ax=ax[0])
ax[0].set_ylabel('Length (cm)')
ax[0].set_title('Sepal Length')

# Boxplot for 'petal.length'
sns.boxplot(x='variety', y='petal.length', data=df, width=0.5, ax=ax[1])
ax[1].set_ylabel('Length (cm)')
ax[1].set_title('Petal Length')

# Adjust layout for better spacing
plt.tight_layout()
plt.suptitle('Boxplots of Sepal and Petal Lengths by Variety',y=1.05)
```

```python
# f) Make a scatter plot of sepal length vs petal length. First all same color, then one color per variety 
df.plot.scatter(x="sepal.length", y="petal.length")


# Add labels and legend
plt.title('Scatter Plot Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.plot()
```

```python
# f) different colors for the varieties
colors = {'Virginica': 'red', 'Setosa': 'blue', 'Versicolor': 'green' }
df.plot.scatter(x="sepal.length", y="petal.length",c=df['variety'].apply(lambda x: colors[x]))

# Add labels to the scatter points for legend
for v, c in colors.items():
    plt.scatter([], [], label=v, color=c)

# Add labels and legend
plt.title('Scatter Plot Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.plot()
```

```python
# f) Easier with seaborn 
scatter = sns.scatterplot(x="sepal.length", y="petal.length", hue="variety",  data=df)

# Add title, display the legend
plt.title('Scatter Plot Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Variety')
```
