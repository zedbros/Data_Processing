```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('../data/medals_long.csv')
df.head()
```

```python
# a) Plot a stacked bar chart of medals counts per nation

# Pivot the DataFrame to get a format suitable for stacked bar chart
# With pivot()
df_crosstab = df.pivot(index='nation', columns='medal', values='count')
# Or with pivot_table
# df_crosstab = pd.pivot_table(df, values='count', index='nation', columns='medal', aggfunc='sum')
# Or with crosstab: 
# df_crosstab = pd.crosstab(index=df['nation'], columns=df['medal'], values=df['count'], aggfunc='sum')


# Plotting the stacked bar chart
ax = df_crosstab.plot.bar(stacked=True)

# Adding labels and title
ax.set_xlabel('Nation')
ax.set_ylabel('Medal Count')
ax.set_title('Stacked Bar Chart of Medal Count by Nation')

# Comment: In practice, we may have to compute the 'count' using e.g. pivot_table or crosstab
```
