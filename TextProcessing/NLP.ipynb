```python
# Read a file
from pathlib import Path
text = Path("res/website-full.txt").read_text()
```

```python
# Number of characters
len(text)
```

```python
import nltk
#Import only needed for the first execusion. 
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

# Number of words
print(len(word_tokenize(text)))

# Number of sentences
number_of_sentences = sent_tokenize(text)
print(len(number_of_sentences))

#number of unique words
print(len(set(word_tokenize(text))))
```

```python
fd = nltk.FreqDist(word_tokenize(text))
print(fd)
print(fd.items())
```

```python
import nltk
from nltk.corpus import stopwords
 
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
 
word_tokens = word_tokenize(text)

# converts the words in word_tokens to lower case and then checks whether 
# they are present in stop_words or not

# filtered_text = [w for w in word_tokens if not w.lower() in stop_words]

#with no lower case conversion
filtered_text = []
 
for w in word_tokens:
    if w not in stop_words:
        filtered_text.append(w)

print(filtered_text)
print(len(filtered_text))
   
print(len(set(filtered_text)))  
```

```python
fd = nltk.FreqDist(filtered_text)

# CREATING A commpressed version
output = ' '.join(fd.keys())
print(output)
```

```python
# and we can derive the table
print(fd.items())
```
