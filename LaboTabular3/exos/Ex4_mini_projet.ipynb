```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
from urllib.request import urlopen
import json
```

```python
with urlopen('https://data.melbourne.vic.gov.au/api/v2/catalog/datasets/live-music-venues/exports/geojson') as response:
    countries = json.load(response)

df = pd.read_csv("gapminder.csv", dtype={"lifeExp": str}).query("year==2007")

fig = px.choropleth(df, geojson=countries, locations="iso_alpha", color="iso_alpha", color_continuous_scale="Viridis", range_color=(0,12), labels={"lifeExp" : "Life expectancy"})
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
