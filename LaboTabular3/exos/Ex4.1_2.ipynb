```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
from urllib.request import urlopen
import json

df = pd.read_csv("gapminder.csv")
```

# 4.1

1.

```python
year = 1997
fig1 = px.treemap(df[df['year']==year], 
                path=[px.Constant("world"), 'continent', 'country'], 
                values='pop',
                color='lifeExp',
                color_continuous_scale='RdBu',
                title=f'Treemap of Life Expectancy in {year}'
                )
fig1.write_html("1997.html")
```

```python
year = 2007
fig2 = px.treemap(df[df['year']==year], 
                path=[px.Constant("world"), 'continent', 'country'], 
                values='pop',
                color='lifeExp',
                color_continuous_scale='RdBu',
                title=f'Treemap of Life Expectancy in {year}'
                )
fig2.write_html("2007.html")
```

```python
df = px.data.gapminder().query("year==1952")
fig = px.scatter_geo(df, locations="iso_alpha", color="continent",
    hover_name="country", size="pop",
    animation_frame="year",
    projection="natural earth")
fig.write_html("4.2.html")
```

### 2.

```python
with urlopen('https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json') as response:
    countries = json.load(response)

df = pd.read_csv("gapminder.csv", dtype={"lifeExp": str}).query("year==2007")

fig = px.choropleth(df, geojson=countries, locations="iso_alpha", color="iso_alpha", color_continuous_scale="Viridis", range_color=(0,12), labels={"lifeExp" : "Life expectancy"})
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
