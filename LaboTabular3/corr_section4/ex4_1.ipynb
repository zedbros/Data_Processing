```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/gapminder.csv')
df.head()
```

```python
# a) Plot and compare two Treemap charts with a life expectancy color map, population rectangle size: in 1952 and in 2007
min_lifeexp = df['lifeExp'].min()
max_lifeexp = df['lifeExp'].max()
df = px.data.gapminder()
year = 2007
fig = px.treemap(df[df['year']==year], 
                  path=[px.Constant("world"), 'continent', 'country'], 
                  values='pop',
                  color='lifeExp',
                  color_continuous_scale='RdBu',
                  range_color=[min_lifeexp,max_lifeexp],
                  title=f'Treemap of life expectancy in {year}'
                  )
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()

```

```python
import plotly.offline as pyo

# Assuming 'df' is your DataFrame
fig = px.treemap(df, path=[px.Constant("world"), 'continent', 'country'], values='pop')

# Save the figure as an HTML file
html_filename = "outputs/treemap.html"
pyo.plot(fig, filename=html_filename, auto_open=False)

# Print the HTML file name
print(f"Treemap saved as {html_filename}")
```
