```python
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('gapminder.csv')
df.head()
```

```python
# a) Plot the gdpPercap in Europe. Make an animation frame to plot it from 1952 to 2007

fig = px.scatter_geo(df, locations="iso_alpha",
                     hover_name="country", size="gdpPercap",
                     animation_frame="year",
                     projection="natural earth",
                     scope='europe',
                     width=1000, height=600,
                     title='Evolution of GDP per Capita in Europe between 1952 and 2007')
fig.show()
```
