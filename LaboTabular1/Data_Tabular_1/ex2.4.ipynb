```python
import numpy as np
import pandas as pd

df = pd.read_csv("iris_cleaned.csv")
```

```python
df.describe()
```

```python
# a)
df["petal_ratio"] = df.apply(lambda i: i["petal.length"]/i["petal.width"], axis=1)
df
```

```python
# b)
numerical_values = ["sepal.length", "sepal.width", "petal.length", "petal.width"]

def z_score_normalization(x, mean, standard_deviation):
    return (x-mean)/standard_deviation

# std_dev = df["sepal.length"].std()
# m = df["sepal.width"].mean()
# print(std_dev, m)
# data = {"index":[]}
st = pd.DataFrame()
for col in numerical_values:
    m = df[col].mean()
    std_dev = df[col].std()
    z = df.apply(lambda i: z_score_normalization(i[col], m, std_dev), axis=1)
    st[f"standardized.{col}"] = z

st
st.pivot_table(index="standardized.sepal.length", values=["standardized.sepal.length", "standardized.sepal.width", "standardized.petal.length", "standardized.petal.width"])
```
