```python
# b)
display(Yangon_city_data.describe())
Yangon_city_data.drop(Yangon_city_data[Yangon_city_data["Rating"] <= 4].index, inplace=True)
Yangon_city_data.drop(Yangon_city_data[9.9 <= Yangon_city_data["Rating"]].index, inplace=True)
display(Yangon_city_data.describe())
```

```python
# c)
Yangon_city_data["New_total"] = Yangon_city_data["Unit price"]*Yangon_city_data["Quantity"] + Yangon_city_data["Tax 5%"]
display(Yangon_city_data)
Yangon_city_data.pivot_table(index=["Invoice ID"], values=["Total", "New_total"])
```

```python
Yangon_city_data["Difference"] = round(Yangon_city_data["Total"] - Yangon_city_data["New_total"])
Yangon_city_data.pivot_table(index=["Invoice ID"], values=["Total", "New_total", "Difference"])
```
