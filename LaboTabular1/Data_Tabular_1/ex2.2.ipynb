```python
import numpy as np
import pandas as pd

df = pd.read_csv("supermarket_sales-Sheet1.csv")
```

```python
df.columns.values
```

```python
df.describe()
```

```python
# a)
s = pd.DataFrame(df.groupby("Payment")["Quantity"].sum())
m = pd.DataFrame(df.groupby("Payment")["Quantity"].mean())
c = pd.DataFrame(df.groupby("Payment")["Quantity"].count())


display("---SUM---", s, "---MEAN---", m, "---COUNT---", c)
```

```python
# b)
# pt = pd.pivot_table(m)
quantity_pivot = df.pivot_table(values='Quantity', index='Payment', aggfunc='mean')
display(quantity_pivot)
```

```python
# c)
customer_pivot = df.pivot_table(values="Quantity", index="Customer type", aggfunc="count")
city_pivot = df.pivot_table(values="Quantity", index="City", aggfunc="count")

display(customer_pivot, city_pivot)
```
