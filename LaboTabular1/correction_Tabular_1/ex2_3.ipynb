```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/supermarket_sales-Sheet1.csv")
df.head()
```

```python
# a) We want to conduct the study only in the Yangon city (for the entire exercise). Filter the data.
print(f"Cities before filtering: {df['City'].unique()}")

# Various ways of doing it. Depending on the task, we may want to create another dataframe e.g. df_yagon
# df = df[df['City'] == 'Yangon'] # Simplest approach
df = df[df['City'].isin(['Yangon'])] # If we wanted to filter multiple cities 
# df = df.loc[df['City'] == 'Yangon'] # More flexible for other filtering
# df = df.loc[df['City'].isin(['Yangon'])] # Similarly, more flexible with multiple cities
# df = df.query('City == "Yangon"') # Express conditions in a more SQL-like syntax.

print(f"Cities after filtering: {df['City'].unique()}")
```

```python
# b) Look at some transactions with low ratings (<=4), some with high rating (>=9.9). Then filter these out for the rest of the exercise
df.loc[df['Rating']<=4].head(10)
```

```python
df.loc[df['Rating']>=9.9].head(10) 
```

```python
df = df.loc[(df['Rating']>4) & (df['Rating']<9.9)]
# or
# df = df.loc[~(df['Rating']<=4) & ~(df['Rating']>=9.9)]
```

```python
# c) Calculate the Total price in a new column, computed as Unit price * quantity + Tax 5%. Compare with the "Total" column
df['Total price'] = df.apply(lambda row: row['Unit price'] * row['Quantity'] + row['Tax 5%'], axis=1)
# or simply
# df['Total price'] = df['Unit price'] * df['Quantity'] + df['Tax 5%']
df[['Total','Total price']].head()
```

```python
# d) A 10% discount on the total price was given to the members. Add that to a new column. Use lambda function.
df['Total price discount'] = df.apply(lambda row: row['Total price']*0.9 if row['Customer type']=='Member' else row['Total price'], axis=1)
df.head()
```
